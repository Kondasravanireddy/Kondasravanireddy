## Using learnt representations
In practice, very few people train an entire Convolutional Network from scratch (with random initialization), because it is relatively rare to have a dataset of sufficient size. Instead, it is common to pretrain a ConvNet on a very large dataset (e.g. ImageNet, which contains 1.2 million images with 1000 categories), and then use the ConvNet either as an initialization or a fixed feature extractor for the task of interest.


We'll train a model to classify ants and bees. We have about 120 training images each for ants and bees. There are 75 validation images for each class. Usually, this is a very small dataset to generalize upon, if trained from scratch. Since we are using transfer learning, we should be able to generalize reasonably well.

This dataset is a very small subset of imagenet.




```python
# Import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, SubsetRandomSampler
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import time
import os
```

```python
# Device configuration (whether to run on GPU or CPU)
device = 'cpu' #torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Download and extract dataset
!wget https://download.pytorch.org/tutorial/hymenoptera_data.zip
!unzip -q hymenoptera_data.zip
```

```python
# Data augmentation and normalization for training
# Just normalization for validation
train_transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
val_transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

data_dir = './hymenoptera_data'
train_dataset = ImageFolder(os.path.join(data_dir, 'train'), train_transform)
val_dataset = ImageFolder(os.path.join(data_dir, 'val'), val_transform)
train_dataloader = DataLoader(train_dataset, batch_size=4,shuffle=True, num_workers=2)
val_dataloader = DataLoader(val_dataset, batch_size=4,shuffle=True, num_workers=2)
class_names = train_dataset.classes
```

```python
def imshow(inp, title=None):
    """Imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)  # pause a bit so that plots are updated


# Get a batch of training data
inputs, classes = next(iter(train_dataloader))

# Make a grid from batch
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

```python
def train_model(model, train_dataloader, criterion, optimizer, num_epochs=25):

    for epoch in range(num_epochs):
        print(f'Epoch {epoch}/{num_epochs - 1}')
        print('-' * 10)
        model.train()
        running_loss = 0.0
        running_corrects = 0

        # Iterate over data.
        for i, (inputs, labels) in enumerate(train_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # zero the parameter gradients
            optimizer.zero_grad()

            # forward
            # track history if only in train
            with torch.set_grad_enabled(True):
              outputs = model(inputs)
              _, preds = torch.max(outputs, 1)
              loss = criterion(outputs, labels)

              # backward + optimize only if in training phase
              loss.backward()
              optimizer.step()

            # statistics
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

        epoch_loss = running_loss / len(train_dataloader.dataset)
        epoch_acc = running_corrects.double() / len(train_dataloader.dataset)

        print(f'Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

        print()
```

```python
def visualize_model(model, num_images=10):
    was_training = model.training
    model.eval()
    images_so_far = 0
    fig = plt.figure()

    with torch.no_grad():
        for i, (inputs, labels) in enumerate(val_dataloader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for j in range(inputs.size()[0]):
                images_so_far += 1
                ax = plt.subplot(num_images//2, 2, images_so_far)
                ax.axis('off')
                ax.set_title(f'predicted: {class_names[preds[j]]}')
                imshow(inputs.cpu().data[j])

                if images_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
from torchvision import models

model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
# Here the size of each output sample is set to 2.
# Alternatively, it can be generalized to nn.Linear(num_ftrs, len(class_names)).
model_ft.fc = nn.Linear(num_ftrs, 2)

model_ft = model_ft.to(device)

loss_func = nn.CrossEntropyLoss()

# Observe that all parameters are being optimized
optimizer = optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 5
train_model(model_ft, train_dataloader, loss_func, optimizer, num_epochs)
```

```python
visualize_model(model_ft)
```

#Answers

1.Data augmentation techniques, such as resizing, cropping, rotating, flipping, or adding noise to training data, are essential in machine learning, particularly in tasks like computer vision. They serve several significant purposes:

Increased Robustness: Augmented data introduces diversity into the training set, making the model more robust to variations in the input data. For instance, if you're training a model to recognize objects in images, different lighting conditions, orientations, and scales are common in real-world scenarios. By augmenting the data with these variations, the model learns to generalize better.

Improved Generalization: Data augmentation helps prevent overfitting by exposing the model to a wider range of scenarios during training. Instead of learning specific details of the training images, the model learns more generalized features that are applicable across different variations of the data. This leads to better performance when the model encounters new, unseen data.

Addressing Class Imbalance: In classification tasks where certain classes are underrepresented, data augmentation can help balance the distribution of classes in the training set. By creating synthetic examples of minority classes through augmentation, the model can learn to classify them more effectively.

Reducing Computational Resource Requirements: Augmenting data allows you to generate a larger effective training set without actually collecting additional data. This can be particularly useful when you have limited labeled data available, as it maximizes the utility of the existing dataset.

Enhancing Performance: When used appropriately, data augmentation can lead to improved model performance, measured in terms of metrics like accuracy, precision, and recall. By exposing the model to a richer set of training examples, it learns to make better predictions on unseen data.

Overall, data augmentation is a powerful technique for enhancing the performance and generalization capabilities of machine learning models, particularly in scenarios where labeled data is limited or when the goal is to build models that can effectively handle real-world variations and challenges.







Answer 2

Training a ResNet model from scratch, without using pre-trained weights, typically requires significantly more data and computational resources compared to fine-tuning a pre-trained model. The performance you achieve depends on various factors such as the size of your dataset, the complexity of the task, the optimization algorithm used, and the training hyperparameters.

When you train a ResNet model from scratch (setting pretrained=False), you start with randomly initialized weights, and the model learns to extract features directly from the raw input data. In general, training from scratch may require more epochs to achieve comparable performance to a pre-trained model, especially for tasks with limited data.

The performance you obtain without using pre-trained weights can vary widely. It's common to observe lower performance compared to using pre-trained weights, especially if you have a small dataset or a complex task. However, with a sufficiently large and diverse dataset, along with careful tuning of hyperparameters, you can still achieve good performance.

To evaluate the performance of a ResNet model trained from scratch, you typically monitor metrics such as accuracy, precision, recall, F1-score, or any other relevant metric for your specific task. These metrics give you insights into how well the model generalizes to unseen data and how effectively it performs the task at hand.







Answer 3

Using a pre-trained ResNet model that was trained on a significantly different dataset may not always lead to good performance when applied directly to a different task, such as classifying ants vs. bees. Here's why:

Domain Shift: If the pre-training dataset is vastly different from the target dataset (ants vs. bees), there might be a domain shift. This means the distribution of features in the pre-training dataset could be quite different from that of the target dataset. As a result, the features learned by the pre-trained model might not be directly applicable to the new task.

Semantic Differences: The semantics of features learned in pre-training might not be relevant to the target task. For example, a pre-trained model trained on a dataset of everyday objects might learn features that are useful for recognizing objects in general scenes but might not capture the specific characteristics of ants and bees.

Overfitting to Pre-training Data: Pre-trained models might overfit to the characteristics of the pre-training dataset, which could limit their generalization to new tasks. Fine-tuning on a new dataset helps mitigate this issue, but if the pre-training dataset is too dissimilar, fine-tuning might not be sufficient.

Task Complexity: The complexity of the target task also plays a role. If the ants vs. bees classification task is significantly different in complexity from the pre-training task, the pre-trained model might not transfer well.

However, despite these challenges, transfer learning with pre-trained models often still provides a good starting point. By fine-tuning the pre-trained model on the target dataset, you can leverage the knowledge encoded in the pre-trained weights and adapt it to the new task. Depending on the similarity between the pre-training and target tasks, fine-tuning may require more or less effort to achieve good performance. Additionally, techniques such as data augmentation, regularization, and adjusting learning rates can help improve the performance of the fine-tuned model.
